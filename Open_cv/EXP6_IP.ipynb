```python
import cv2 as cv
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

```python
j=cv.imread('/content/download.jpg')
```

```python
grey_image=cv.cvtColor(j,cv.COLOR_RGB2GRAY)
```

```python
cv2_imshow(grey_image)
```

```python
blur_IMAGE=cv.GaussianBlur(grey_image,(9,9),1.0)
```

```python
cv2_imshow(blur_IMAGE)
cv.imwrite('/content/Images/GaussianBlure.jpg',blur_IMAGE)
```

```python
blur=cv.blur(grey_image,(9,9),10)
cv2_imshow(blur)
cv.imwrite('/content/Images/blur.jpg',blur)
```

```python
median_blur=cv.medianBlur(grey_image,9,11)
cv2_imshow(median_blur)
cv.imwrite('/content/Images/median_blur.jpg',median_blur)
```

```python
blur_bilateral = cv.bilateralFilter(grey_image,9,75,75)
cv2_imshow(blur_bilateral)
cv.imwrite('/content/Images/blur_bilateral.jpg',blur_bilateral)
```
