```python
# 1. Two Sum
# Easy
# Topics
# premium lock icon
# Companies
# Hint
# Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

# You may assume that each input would have exactly one solution, and you may not use the same element twice.

# You can return the answer in any order.

# Example 1:

# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]
# Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
# Example 2:

# Input: nums = [3,2,4], target = 6
# Output: [1,2]
# Example 3:

# Input: nums = [3,3], target = 6
# Output: [0,1]
 

# Constraints:

# 2 <= nums.length <= 104
# -109 <= nums[i] <= 109
# -109 <= target <= 109
# Only one valid answer exists.
 

# Follow-up: Can you come up with an algorithm that is less than O(n2) time complexity?
```

```python
nums = list(map(int, input("Enter the input list: ").split()))
target = int(input("Enter the target value: "))

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == target:
            print([i, j])
```

```python
# Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.

# The overall run time complexity should be O(log (m+n)).

 

# Example 1:

# Input: nums1 = [1,3], nums2 = [2]
# Output: 2.00000
# Explanation: merged array = [1,2,3] and median is 2.
# Example 2:

# Input: nums1 = [1,2], nums2 = [3,4]
# Output: 2.50000
# Explanation: merged array = [1,2,3,4] and median is (2 + 3) / 2 = 2.5.
```

```python
num1 = list(map(int, input("Enter the input list: ").split()))
num2 = list(map(int, input("Enter the input list: ").split()))

merged_list=num1+num2

merged_list.sort()
n = len(merged_list)
if len(merged_list)%2==0:
    value = (merged_list[n//2 - 1] + merged_list[n//2]) / 2
else:
    value = merged_list[n//2]
print(value)
```

```python
# Longest Palindromic Substring
# Medium
# Topics
# premium lock icon
# Companies
# Hint
# Given a string s, return the longest palindromic substring in s.

 

# Example 1:

# Input: s = "babad"
# Output: "bab"
# Explanation: "aba" is also a valid answer.
# Example 2:

# Input: s = "cbbd"
# Output: "bb"
```

```python
string=input(' Enter a string without symbols')
List=list(string)
startpos=0
endpos=0
for i in range(len(List)):
    current_element=str(List[i])
    for j in range(i+1,len(List)):
        if current_element==List[j]:
            startpos=i
            endpos=j
    
print(''.join(List[startpos:endpos+1]))       
```
