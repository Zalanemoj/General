```python
A=[3,6,9,12,15]
B=[4,8,12,16,20]
C=[6.7,8.6,12.2,14.3,19]
plt.scatter(A,B)
plt.scatter(A,C)
plt.plot(A,B, color="r",label="Reference")
plt.plot(A,C, color="g",label="Practical")
plt.ylabel("Current mA")
plt.xlabel("Pressure psi")
plt.legend()
plt.title("Pressure to Current")
plt.savefig("P to I.png")
plt.show()
```
